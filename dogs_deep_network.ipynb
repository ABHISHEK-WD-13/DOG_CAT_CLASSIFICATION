```python
import os
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
image_size = 64
```

```python
dir = "archive/dataset/training_set/"  
catagories = ["cats" , "dogs"]
train_data = []
for categ in catagories:
 try:
     path = os.path.join(dir , categ)  
     class_name = catagories.index(categ)
     for img in os.listdir(path):
          img_array = cv2.imread(os.path.join(path, img) , cv2.IMREAD_GRAYSCALE)
          new_array = cv2.resize(img_array , (image_size , image_size))
          train_data.append([new_array , class_name])
 except Exception as e:
     pass
```

```python
import random
random.shuffle(train_data)
data = np.array(train_data)
# print(data[:][1])
y_train=np.zeros((1,len(train_data)))
x_train_t=np.zeros((len(train_data), image_size*image_size))
# print(x_train_t[1].shape)
for i in range(len(train_data)):
    image_array = data[i][0]
    image_array = image_array.reshape(1,image_size*image_size )
    x_train_t[i]= image_array
    y_train[0][i] = data[i][1]
x_train = x_train_t.T/255
```

```python
dir_test = "archive/dataset/test_set/"
catagorie = ["cats" , "dogs"]
test_data = []
for c in catagorie:
 try:
     path_test = os.path.join(dir_test , c)  
     class_name_test = catagorie.index(c)
     for img_test in os.listdir(path_test):
          img_array_test = cv2.imread(os.path.join(path_test, img_test) , cv2.IMREAD_GRAYSCALE)
          new_array_test = cv2.resize(img_array_test , (image_size , image_size))
          test_data.append([new_array_test , class_name_test])
 except Exception as e_test:
     pass
print(test_data)
random.shuffle(test_data)

data_test = np.array(test_data)
# print(data[:][1])
y_test=np.zeros((1,len(test_data)))
x_test_t=np.zeros((len(test_data), image_size*image_size))
# print(x_train_t[1].shape)
for i in range(len(test_data)):
    image_array_test = data[i][0]
    image_array_test = image_array_test.reshape(1,image_size*image_size )
    x_test_t[i]= image_array_test
    y_test[0][i] = data_test[i][1]
x_test = x_test_t.T/255
```

```python
#INITIALIZE PARAMETERS

def initialize_params(size_input , size_hidden , size_output):
    W1 = np.random.randn(size_hidden , size_input)*np.sqrt(1/(size_input))
    b1 = np.zeros((size_hidden , 1))
    W2 = np.random.randn(size_output , size_hidden)*np.sqrt(1/(size_input))
    b2 = np.zeros((size_output , 1))
    params = {"W1":W1,
              "W2":W2,
              "b1":b1,
              "b2":b2}
    return params
```

```python


def sigmoid(z):
    s=1/(1+np.exp(-z))
    return s

def linear_forward(X,W_1,W_2,b_1,b_2):
    Z1 = np.dot(W_1,X)+b_1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W_2,A1)+b_2
    A2 = sigmoid(Z2)
    return A2,A1,Z1,Z2
```

```python
def cost(A2 , y):
    m = y.shape[1]
    computed_cost = (-1/m) * (np.dot(y, np.log(A2).T) + np.dot((1-y), np.log(1-A2).T))
    computed_cost= np.squeeze(computed_cost)      
    return computed_cost
```

```python
def back_prop(A2,A1,y,W2,Z1,Z2,x,t):
    m=y.shape[1]
    dZ2=A2-y
    dW2=(1/m)*np.dot(dZ2 , A1.T)
    db2=(1/m)*np.sum(dZ2 , axis=1,keepdims=True)
    dZ1=np.dot(W2.T , dZ2)*(A1-np.power(A1,2))
    dW1=(1/m)*np.dot(dZ1,x.T)
    db1=(1/m)*np.sum(dZ1,axis=1,keepdims=True)
    beta=0.9
    beta2=0.999
    vdW1=np.zeros(dW1.shape)
    vdb1=np.zeros(db1.shape)
    vdW2=np.zeros(dW2.shape)
    vdb2=np.zeros(db2.shape)

    sdW1=np.zeros(dW1.shape)
    sdb1=np.zeros(db1.shape)
    sdW2=np.zeros(dW2.shape)
    sdb2=np.zeros(db2.shape)
    
    vdW1=beta*vdW1+(1-beta)*dW1
    vdW1_c=vdW1/(1-np.power(beta,t))

    vdW2=beta*vdW2+(1-beta)*dW2
    vdW2_c=vdW2/(1-np.power(beta,t))
    
    vdb1=beta*vdb1+(1-beta)*db1
    vdb1_c=vdb1/(1-np.power(beta,t))
    
    vdb2=beta*vdb2+(1-beta)*db2
    vdb2_c=vdb2/(1-np.power(beta,t))
    
    sdW1=beta2*sdW1+(1-beta2)*np.power(dW1,2)
    sdW1_c=sdW1/(1-np.power(beta2,t))
    
    sdW2=beta2*sdW2+(1-beta2)*np.power(dW2,2)
    sdW2_c=sdW2/(1-np.power(beta2,t))
    
    sdb1=beta2*sdb1+(1-beta2)*np.power(db1,2)
    sdb1_c=sdb1/(1-np.power(beta2,t))
    
    sdb2=beta2*sdb2+(1-beta2)*np.power(db2,2)
    sdb2_c=sdb2/(1-np.power(beta2,t))
    
    grads={
        "vdW1":vdW1_c,
        "vdW2":vdW2_c,
        "vdb1":vdb1_c,
        "vdb2":vdb2_c,
        "sdW1":sdW1_c,
        "sdW2":sdW2_c,
        "sdb1":sdb1_c,
        "sdb2":sdb2_c
    }
    return grads
```

```python
def update_params(W1,W2,b1,b2,grads,learning_rate):
    vdW1=grads["vdW1"]
    vdW2=grads["vdW2"]
    vdb1=grads["vdb1"]
    vdb2=grads["vdb2"]
    sdW1=grads["sdW1"]
    sdW2=grads["sdW2"]
    sdb1=grads["sdb1"]
    sdb2=grads["sdb2"]
    e=0.00000001
    W1=W1-learning_rate*(vdW1/(np.sqrt(sdW1+e)))
    W2=W2-learning_rate*(vdW2/(np.sqrt(sdW2+e)))
    b1=b1-learning_rate*(vdb1/(np.sqrt(sdb1+e)))
    b2=b2-learning_rate*(vdb2/(np.sqrt(sdb2+e)))
    params = {"W1":W1,
              "W2":W2,
              "b1":b1,
              "b2":b2}
    return params

    
```

```python
def model(x,y,learning_rate=0.005,iterations=1000,neuron=0):
    params = initialize_params(x.shape[0] , neuron , 1)
    t=0
    W1 = params["W1"]
    W2 = params["W2"]
    b1 = params["b1"]
    b2 = params["b2"]   
    for i in range(iterations):
       A2,A1,Z1,Z2 = linear_forward(x,W1,W2,b1,b2)
       computed_cost=cost(A2,y)
    #    computed_cost_test=cost(A2,y)
       t=t+1
       grads=back_prop(A2,A1,y,W2,Z1,Z2,x,t)
       params_updated=update_params(W1,W2,b1,b2,grads,learning_rate)
       W1 = params_updated["W1"]
       W2 = params_updated["W2"]
       b1 = params_updated["b1"]
       b2 = params_updated["b2"] 
    #    if i%100==0:
    #    print(str(computed_cost)+"b2"+str(b2)+str(i))
      
    return (W1,W2,b1,b2)   
```

```python
def predict(X,W_1,W_2,b_1,b_2):
    m=X.shape[1]
    Z1 = np.dot(W_1,X)+b_1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W_2,A1)+b_2
    A2 = sigmoid(Z2)
    y_predicted = np.zeros((1,m))
    for i in range(A2.shape[1]):
        if A2[0][i] > 0.5 :
            y_predicted[0][i]=1
        else :
            y_predicted[0][i]=0

    return y_predicted        
```

```python
for n in range(100):
    (W_1,W_2,b_1,b_2) = model(x_train,y_train,0.005,20,n)
    train_ans=predict(x_train,W_1,W_2,b_1,b_2)
    print("train accuracy: {} %".format(100 - np.mean(np.abs(train_ans - y_train)) * 100))
    test_ans=predict(x_test,W_1,W_2,b_1,b_2)
    print("test accuracy: {} %".format(100 - np.mean(np.abs(test_ans - y_test)) * 100))
    print("b2"+str(b_2))
```
