```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import cv2
```

```python
image_size=64
dir = "archive/dataset/training_set/"  
catagories = ["cats" , "dogs"]
train_data = []
for categ in catagories:
 print(categ) 
 try:
     path = os.path.join(dir , categ)  
     class_name = catagories.index(categ)
     for img in os.listdir(path):
          img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
          new_array = cv2.resize(img_array , (image_size , image_size))
          train_data.append([new_array , class_name])
 except Exception as e:
     pass
```

```python
import random
random.shuffle(train_data)
# print(train_data[1][0])
x_train = np.zeros((len(train_data),64,64,3))
y_train = np.ndarray((len(train_data),) , int)
for i in range(len(train_data)):
    x_train[i]=train_data[i][0]
    y_train[i]=train_data[i][1]
```

```python

# data = np.array(train_data)
# print(data[:][1])
# y_train=np.zeros((1,len(train_data)))
# x_train_t=np.zeros((len(train_data), image_size*image_size*3))
# print(x_train_t[1].shape)
# for i in range(len(train_data)):
    # image_array = data[i][0]
    # image_array = image_array.reshape(1,image_size*image_size*3 )
    # x_train_t[i]= image_array
    # y_train[0][i] = data[i][1]
x_train = x_train/255
```

```python
dir_test = "archive/dataset/test_set/"
catagorie = ["cats" , "dogs"]
test_data = []
for c in catagorie:
 try:
     path_test = os.path.join(dir_test , c)  
     class_name_test = catagorie.index(c)
     for img_test in os.listdir(path_test):
          img_array_test = cv2.imread(os.path.join(path_test, img_test) , cv2.IMREAD_COLOR)
          new_array_test = cv2.resize(img_array_test , (image_size , image_size))
          test_data.append([new_array_test , class_name_test])
 except Exception as e_test:
     pass
# print(test_data)

# data_test = np.array(test_data)
# print(data[:][1])
# y_test=np.zeros((1,len(test_data)))
# x_test_t=np.zeros((len(test_data), image_size*image_size*3))
# print(x_train_t[1].shape)
# for i in range(len(test_data)):
    # image_array_test = data[i][0]
    # image_array_test = image_array_test.reshape(1,image_size*image_size*3)
    # x_test_t[i]= image_array_test
    # y_test[0][i] = data_test[i][1]
random.shuffle(test_data)
x_test = np.zeros((len(test_data),64,64,3))
y_test = np.ndarray((len(test_data),) , int)
for i in range(len(test_data)):
    x_test[i]=test_data[i][0]
    y_test[i]=test_data[i][1]
x_test = x_test/255
```

```python
from tensorflow import keras
#kernel_regularizer=tf.keras.regularizers.L2(0.0001)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(64,64,3)),
                            tf.keras.layers.Dense(256,activation='relu'),
                            tf.keras.layers.Dropout(0.25),
                            tf.keras.layers.Dense(128,activation='relu'),
                            tf.keras.layers.Dropout(0.25),
                            tf.keras.layers.Dense(64,activation='relu'),
                            tf.keras.layers.Dropout(0.25),
                            tf.keras.layers.Dense(25,activation='relu'),
                            tf.keras.layers.Dropout(0.25),
                            tf.keras.layers.Dense(12,activation='relu'),
                            tf.keras.layers.Dropout(0.25),
                            # activity_regularizer=tf.keras.regularizers.L2(0.01)),
                            tf.keras.layers.Dense(12,activation='relu'),
                            # activity_regularizer=tf.keras.regularizers.L2(0.01)),
                            tf.keras.layers.Dropout(0.25),
                            # tf.keras.layers.Dense(6,activation='relu'),
                            # tf.keras.layers.Dense(6,activation='relu'),
                            tf.keras.layers.Dense(1,activation="sigmoid")])
```

```python
model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
```

```python
model.fit(x_train , y_train , epochs=50)
```

```python
loss_test , acc_test = model.evaluate(x_test,y_test , verbose=2)
print(loss_test , acc_test)
```
